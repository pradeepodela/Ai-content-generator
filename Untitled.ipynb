```python
def get_topnews(country='IN',period='10d',results=25):
    google_news = GNews(language='en', country=country, period='7d', max_results=results, exclude_websites=['youtube.com'])
    news = google_news.get_top_news()
    return news
```

```python
print(get_topnews(country='IN',period='10d',results=3))
```
